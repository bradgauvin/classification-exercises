1. Attendance Data
- Read the data from the attendance table and calculate an attendance percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
import pandas as pd
import numpy as np
import seaborn as sns
from pydataset import data
```

```python
df = pd.read_csv('attendance.csv')
df
```

```python
df.shape
```

```python
df_melt = df.melt(id_vars = 'Unnamed: 0',
                          var_name = 'day',
                        value_name = 'value')
df_melt
```

```python
df_melt['attendance_value'] = df_melt.value.map({'A':0, 'T':.9, 'P': 1, 'H':.5})
df_melt
```

```python
df_tidy = df_melt.groupby('Unnamed: 0').mean('attendance_value')
df_tidy
```

3. Cake Recipes
- Read the cake_recipes table. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.
- Tidy the data as necessary.
- Which recipe, on average, is the best? recipe b
- Which oven temperature, on average, produces the best results? 275
- Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
data = pd.read_csv('cake_recipes.csv')
data
```

```python
data.info()
```

```python
data[['recipe', 'position']] = data['recipe:position'].str.split(':', expand = True)
data = data.drop(columns = ['recipe:position'])
data
```

```python
data_melt = data.melt(id_vars = ['recipe', 'position'],
         var_name = 'temp',
         value_name = 'score')
data_melt
```

```python
# Which recipe, on average, is the best? recipe b
data_melt.groupby('recipe').score.mean()
```

```python
# Which oven temperature, on average, produces the best results? 275
data_melt.groupby('temp').score.mean()
```

```python
# Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees
data_melt.sort_values(by='score').tail(1)
```
